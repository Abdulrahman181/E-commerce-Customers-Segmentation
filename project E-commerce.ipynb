## Applying KMeans Model

```python
df['Number of Transactions'] = df.groupby('customer_id')['transaction_id'].transform('count')
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
label_encoders = {}
for column in ['transaction_status', 'coupon_name']: 
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column].astype(str))
    label_encoders[column] = le
```

```python
df['Number of Transactions'] = df.groupby('customer_id')['transaction_id'].transform('count')
features = ['gender_id', 'city_id', 'transaction_status', 'coupon_name', 'Number of Transactions']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[features])
```

```python
sse = []
silhouette_scores = []

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))
```

```python
optimal_k = silhouette_scores.index(max(silhouette_scores)) + 2
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(X_scaled)
df['cluster'] = kmeans.labels_
```

```python
numeric_columns = df.select_dtypes(include=['number']).columns
cluster_analysis = df.groupby('cluster')[numeric_columns].mean()
print(cluster_analysis)
```

```python
X = df[features]
kmeans = KMeans(n_clusters=10, random_state=42, n_init=20)
clusters = kmeans.fit_predict(X_scaled)
silhouette_avg = silhouette_score(X, clusters)
print(f"Silhouette Score: {silhouette_avg:.2f}")
```

```python
best_n_clusters = 0
best_score = -1

for n_clusters in range(2, 11):  
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    clusters = kmeans.fit_predict(X_scaled)
    silhouette_avg = silhouette_score(X_scaled, clusters)
    print(f"Number of clusters: {n_clusters}, Silhouette Score: {silhouette_avg:.2f}")
    
    if silhouette_avg > best_score:
        best_score = silhouette_avg
        best_n_clusters = n_clusters

print(f"Best number of clusters: {best_n_clusters}, with Silhouette Score: {best_score:.2f}")
```

```python
kmeans = KMeans(n_clusters=3, random_state=42) 
df['Cluster'] = kmeans.fit_predict(X)
plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x='gender_id', y='city_id', hue='Cluster', palette='viridis', marker='o')
plt.title('Data After K-Means')
plt.xlabel('Gender ID')
plt.ylabel('City ID')
plt.tight_layout()
plt.show()
```

## Applying DBSCAN Model

```python
from sklearn.cluster import DBSCAN
```

```python
features = ['gender_id', 'city_id'] 
X = df[features]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)  
df['Cluster'] = dbscan.fit_predict(X_scaled)
n_clusters = len(set(df['Cluster'])) - (1 if -1 in df['Cluster'].values else 0)
n_noise = list(df['Cluster']).count(-1)
print(f"Number of clusters: {n_clusters}")
print(f"Number of noise points: {n_noise}")
```

```python
plt.subplot(1, 2, 2)
sns.scatterplot(data=df, x='gender_id', y='city_id', hue='Cluster', palette='viridis', marker='o', legend='full')
plt.title('Data After DBSCAN')
plt.xlabel('Gender ID')
plt.ylabel('City ID')
plt.tight_layout()
plt.show()
```

```python
X_no_noise = X_scaled[df['Cluster'] != -1]
clusters_no_noise = df['Cluster'][df['Cluster'] != -1]
silhouette_avg = silhouette_score(X_no_noise, clusters_no_noise)
print(f"Silhouette Score: {silhouette_avg:.2f}")
```
